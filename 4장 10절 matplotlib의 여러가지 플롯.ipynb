# 바 차트
### 1. bar
### 2. barh : 가로 방향으로 바 차트를 그리는 경우

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font', family='12LotteMartDreamMedium')
mpl.rc('axes', unicode_minus=False)
```

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = [u'가', u'나', u'다']
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.show()
```

### xerr 인수나 yerr 인수를 지정하면 에러 바(error bar)를 추가할 수 있다. ### alpha는 투명도를 지정한다. 0이면 완전 투명, 1이면 완전 불투명이다.

```python
people = (u'ppp', u'jjj', u'jyh', u'ccc', u'ccm')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.barh(y_pos, performance, xerr=error, align='center', alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel(u'x 라벨')
plt.show()
```

```python
n_groups = 5
means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)
means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, means_men, bar_width,
                 alpha=opacity,
                 color='b',
                 yerr=std_men,
                 error_kw=error_config,
                 label='Men',
                 align='edge')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 alpha=opacity,
                 color='r',
                 yerr=std_women,
                 error_kw=error_config,
                 label='Women',
                 align='edge')

plt.xlabel('Group')
plt.ylabel('Sources')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E'))
plt.legend()
plt.tight_layout()
plt.show()
```

#### bottom 인수 : 바의 위치를 조정하여 겹친 바 차트(stacked bar chart)도 그릴 수 있다.

```python
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind, menMeans, width, color='r', yerr=menStd)
p2 = plt.bar(ind, womenMeans, width, color='y', bottom=menMeans, yerr=womenStd)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1','G2','G3','G4','G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
plt.show()
```

# 스템 플롯
### 바 차트와 유사하지만 폭(width)이 없는 스템 플롯(stem plot)도 있다
### 이산 확률 함수나 자기상관관계(auto-correlation) 묘사 시 주로 사용

```python
x = np.linspace(0.1, 2*np.pi, 10)
markerline, stemlines, baseline = plt.stem(x, np.cos(x), '-.')
plt.setp(markerline, 'markerfacecolor', 'b')
plt.setp(baseline, 'color', 'r', 'linewidth', 2)
plt.show()
```

# Imshow
### 행과 열을 가진 2차원 데이터의 시각화 / 화상(image) 데이터 : 전형적 2차원 자료

```python
from sklearn.datasets import load_digits
digits = load_digits()
x = digits.images[0]
x
```

```python
plt.imshow(x, interpolation='nearest')
plt.grid(False)
plt.show()
```

```python
methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
           'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']
fig, axes = plt.subplots(3, 6, figsize=(12, 6), subplot_kw={'xticks': [], 'yticks': []})
fig.subplots_adjust(hspace=0.3, wspace=0.05)
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(x, interpolation=interp_method)
    ax.set_title(interp_method)
plt.show()
```

# 컨투어 플롯
### 입력 변수가 x,y 두개, 출력 변수가 z 하나인 경우 -> 3차원 자료
### 3차원 자료를 시각화 하는 방법 -> 등고선(contour)을 사용

```python
def f(x, y):
    return (14-x/2+x**5+y**3)*np.exp(-x**2-y**2)
n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
xx, yy = np.meshgrid(x, y)
zz = f(xx, yy)
```

```python
plt.contourf(xx, yy, zz, alpha=.75, cmap='jet')
plt.contour(xx, yy, zz, colors='black', linewidth=.5)
plt.show()
```

# 3D 서비스 플롯

```python
from mpl_toolkits.mplot3d import Axes3D
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
xx, yy = np.meshgrid(x, y)
rr = np.sqrt(xx**2 + yy**2)
zz = np.sin(rr)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(xx, yy, zz, rstride=1, cstride=1,cmap='hot')
plt.show()
```
