# 틱 설정 
## 틱(tick) : 플롯이나 차트에서 축상의 위치 표시 지점
## xticks or yticks 명령 사용
## 틱 라벨 문자열에는 $$사이에 LateX 수학 문자식 넣을 수 있음

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.show()
```

# 범례
### legend, 범례의 위치는 자동으로 정해짐
### 수동설정 시 : loc 인수 사용

```python
X=np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.plot(X, S, label="sins")
plt.legend(loc=2)
plt.show()
```

# 2. Axes와 Subplot
#### - 하나의 윈도우(Figure)안에 여러개의 플롯을 배열 형태로 보여야 하는 경우도 있다.
#### - Figure 안에 있는 각각의 플롯은 Axes라고 불리는 객체에 속한다.
#### - subplot : 명시적으로 Axes 객체를 얻기 위한 명령어. plot명령을 바로 사용해도 자동으로 Axes 생성
### - subplot : 그리드 형태의 Axes 객체들을 생성. Figure가 행렬(matrix)이고 Axes가 행렬의 원소라 생각하면 됨.
### - subplot : 세개의 인수를 가지는데 처음 두개의 원소가 전체 그리드 행렬의 모양을 지시하는 두 숫자, 세번째 인수가 네 개 중 어느것인지를 의미하는 숫자

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 suplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.show()
```

```python
plt.subplot(221); plt.plot([1, 2]); plt.title(1)
plt.subplot(222); plt.plot([1, 2]); plt.title(2)
plt.subplot(223); plt.plot([1, 2]); plt.title(3)
plt.subplot(224); plt.plot([1, 2]); plt.title(4)
plt.tight_layout()
plt.show()
```

```python
plt.title(u'한글 제목')
plt.plot(X, C, label=u"코사인")
t = 2 * np.pi / 3
plt.scatter(t, np.cos(t), 50, color='blue')
plt.xlabel(u"엑스축 라벨")
plt.ylabel(u"와이축 라벨")
plt.annotate(u"여기가 0.5!", xy=(t, np.cos(t)), xycoords='data', xytext=(-90, -50), 
             textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->"))
plt.show()  
```

#### 개별적으로 폰트 적용 : fontdict인수에 넣는다. 

```python
font1 = {'family': '12LotteMartDreamBold', 'color': 'black', 'size': 24}
font2 = {'family': 'HYGothic-Medium', 'color': 'darkred', 'weight': 'normal', 'size': 18}
font3 = {'family': 'HYGungSo-Bold', 'color': 'blue', 'weight': 'light', 'size': 12}

x = np.linspace(0.0, 5.0, 100)
y = np.cos(2*np.pi*x)*np.exp(-x)

plt.plot(x, y, 'k')
plt.title(u'한글 제목', fontdict=font1)
plt.xlabel(u'엑스 축', fontdict=font2)
plt.ylabel(u'와이 축', fontdict=font3)
plt.subplots_adjust()
plt.show()
```
