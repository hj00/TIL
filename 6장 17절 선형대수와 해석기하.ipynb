# 벡터의 기하학적 의미
### N차원 벡터 a는 N차원의 공간상에서
#### 1. 벡터 a의 값으로 표시되는 점(point) 또는
#### 2. 원점과 벡터 a의 값으로 표시되는 점을 연결한 화살표(arrow)로 간주할 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1, 2])
a
```

```python
plt.annotate('', xy=a, xytext=(0,0), arrowprops=dict(facecolor='black'))
plt.plot(0, 0, 'ro', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.text(0.35, 1.15, "$a$", fontdict={"size": 18})
plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-2.4, 3.4)
plt.ylim(-1.2, 3.2)
plt.grid(True)
plt.show()
```

# 벡터의 길이
#### 피타고라스 정리를 이용하여 놈(norm) ||a||로 계산 가능
#### ||a|| = sqrt(a.T, a)=sqrt(a1^2 + ... + aN^2)
#### norm명령

```python
a = np.array([1, 2])
np.linalg.norm(a)
```

# 단위 벡터
### 길이가 1인 벡터
### 임의의 벡터 x에 대해 x/||x|| -> 단위벡터

```python
a = np.array([1, 0])
b = np.array([0, 1])
c = np.array(1/np.sqrt(2), 1/np.sqrt(2))
np.linalg.norm(a), np.linalg.norm(b), np.linalg.norm(c)
```

# 벡터와 실수의 곱
### 벡터와 양의 실 수 곱 -> 벡터 방향 그대로, 실수의 크기만큼 벡터의 길이가 커진다
### 벡터와 음의 실수 곱 -> 벡터의 방향 반대.

```python
a = np.array([1, 2])
b = 2 * a
c = -1 * a

plt.plot(0,0,'ro',ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.annotate('', xy=b, xytext=(0,0), arrowprops=dict(facecolor='black'))
plt.text(0.35, 1.15, "$a$", fontdict={"size": 18})

plt.plot(0,0,'ro',ms=10)
plt.plot(c[0], c[1], 'ro', ms=10)
plt.annotate('', xy=c, xytext=(0,0), arrowprops=dict(facecolor='black'))

plt.xticks(np.arange(-5, 6))
plt.yticks(np.arange(-5, 6))
plt.xlim(-4.4, 5.4)
plt.ylim(-3.2, 5.2)
plt.grid(True)
plt.show()
```

# 벡터의 합
### 벡터와 벡터의 합도 벡터
### 벡터의 합으로 만들어진 벡터의 위치 : 두 벡터를 이웃하는 변으로 가지는 평행사변형의 다른 쪽 모서리의 위치

```python
a = np.array([1, 2])
b = np.array([2, 1])
c = a + b

plt.annotate('', xy=a, xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=b, xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=c, xytext=(0,0), arrowprops=dict(facecolor='black'))

plt.plot(0, 0, 'ro', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)
plt.plot(c[0], c[1], 'ro', ms=10)
plt.plot([a[0], c[0]], [a[1], c[1]], 'k--')
plt.plot([b[0], c[0]], [b[1], c[1]], 'k--')

plt.text(0.35, 1.15, "$a$", fontdict={"size": 18})
plt.text(1.15, 0.25, "$b$", fontdict={"size": 18})
plt.text(1.25, 1.45, "$c$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)
plt.grid(True)
plt.show()
```

# 벡터의 차
### a - b = c : 벡터 b가 가ㅣ키는 점으로부터 벡터 a가 가리키는 점을 연결하는 벡터
### 벡터 b에 벡터 a - b를 더하면 즉, b와 벡터 a-b를 연결도록 벡터 a가 도어야 하기 때문
#### a -b=c
#### b+c=b+(a-b)=a

```python
a = np.array([1, 2])
b = np.array([2, 1])
c = a - b

plt.annotate('', xy=a, xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=b, xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=a, xytext=b, arrowprops=dict(facecolor='black'))

plt.plot(0, 0, 'ro', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)

plt.text(0.35, 1.15, "$a$", fontdict={"size": 18})
plt.text(1.15, 0.25, "$b$", fontdict={"size": 18})
plt.text(1.55, 1.65, "$a-b$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 5))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.4, 4.4)
plt.ylim(-0.6, 3.8)
plt.grid(True)
plt.show()
```

# 투영
### 벡터 a -> 다른 벡터 b에 직교하는 성분 a1, 벡터 b에 평행한 성분 a2로 분해 가능
### 성분 a2의 길이를 벡터 a의 벡터 b에 대한 투영(projection)이라고 한다.
#### a = a1 + a2 / a1 ㅗ b and a2 = a-a1
#### ||a2||  =  a^T*(b/||b||) = a^Tb/||b|| = b^Ta/||b||

```python
a = np.array([1, 2])
b = np.array([2, 0])
a2 = np.dot(a, b)/np.linalg.norm(b)*np.array([1, 0])
a1 = a - a2

plt.annotate('', xy=a, xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=b, xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=a2, xytext=(0,0), arrowprops=dict(facecolor='green'))
plt.annotate('', xy=a1, xytext=(0,0), arrowprops=dict(facecolor='green'))

plt.plot(0, 0, 'ro', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)
plt.plot(b[0], b[1], 'ro', ms=10)

plt.text(0.35, 1.15, "$a$", fontdict={"size": 18})
plt.text(1.55, 0.15, "$b$", fontdict={"size": 18})
plt.text(-0.35, 1.05, "$a_1$", fontdict={"size": 18})
plt.text(0.50, 0.15, "$a_2$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.5, 3.5)
plt.ylim(-0.5, 3)
plt.grid(True)
plt.show()
```

# 직선의 방정식
#### f(x)  =  w^T(x-w)  =  w^Tx-w^Tw = w^Tx-||w||^2  = w^Tx - w0 = 0
### x : 직선 상의 점을 나타내는 벡터, w : 원점으로 직선까지 이어지는 수직선을 나타내는 벡터 / x-w 벡터가 w 벡터와 수직 -> x가 가리키는 점과 w가 가리키는 점을 이은 선이 w와 수직이라는 뜻

```python
w = np.array([1, 2])
x1 = np.array([3, 1])
x2 = np.array([-1, 3])
w0 = 5

plt.annotate('', xy=w, xytext=(0,0), arrowprops=dict(facecolor='red'))
plt.annotate('', xy=x1, xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=x2, xytext=(0,0), arrowprops=dict(facecolor='gray'))

plt.plot(0, 0, 'ro', ms=10)
plt.plot(w[0], w[1], 'ro', ms=10)
plt.plot(x1[0], x1[1], 'ro', ms=10)
plt.plot(x2[0], x2[1], 'ro', ms=10)
plt.plot([-3, 5], [4, 0], 'r-', lw=5)

plt.text(0.35, 1.15, "$w$", fontdict={"size": 18})
plt.text(1.55, 0.25, "$x_1$", fontdict={"size": 18})
plt.text(-0.9, 1.40, "$x_2$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.7, 4.2)
plt.ylim(-0.5, 3.5)
plt.grid(True)
plt.show()
```

# 직선과 점의 거리
#### w^Tx - w0 = 0 과 이 직선 위에 있지 않은 점 x'의 거리는 단위벡터 w/||w||에 대한 x'의 투영에서 ||w||를 뺀 값의 절대값

# 벡터의 선형 종속과 선형 독립
### 벡터들의 선형 조합이 0이 되는 모두 0이 아닌 스칼라값들이 존재하면 그 벡터들은 선형 종속(linearly dependent)
### 벡터들의 선형 조합이 0이 되는 모두 0이 아닌 스칼라값들이 존재하지 않으면 그 벡터들은 선형 독립(linearly independent)
#### alpha1a1 + alpha2a2 + ... + alphaKaK = 0 <-> alpha1 = ... = alphaK = 0

```python
a = np.array([1, 2])
b = np.array([3, 3])
c = np.array([10, 14])
2 * a + b - 0.5 * c
```

# 좌표(coordinate)
### 좌표는 표준 기저 벡터들에 대한 그 벡터의 투영(projection)으로 볼 수 있다.
#### a = a1*e1 + a2*e2(a1, a2는 좌표라는 스칼라 값, e1,e2는 벡터)

```python
e1 = np.array([1, 0])
e2 = np.array([0, 1])
a = np.array([2, 2])

plt.annotate('', xy=e1, xytext=(0,0), arrowprops=dict(facecolor='green'))
plt.annotate('', xy=e2, xytext=(0,0), arrowprops=dict(facecolor='green'))
plt.annotate('', xy=a, xytext=(0,0), arrowprops=dict(facecolor='gray'))

plt.plot(0, 0, 'ro', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)

plt.text(1.05, 1.35, "$a$", fontdict={"size": 18})
plt.text(-0.3, 0.5, "$e_2$", fontdict={"size": 18})
plt.text(0.58, -0.2, "$e_1$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.5, 3.5)
plt.ylim(-0.5, 3)
plt.grid(True)
plt.show()
```

# 변환행렬(Transform Matrix)
### 원래의 기저벡터가 아닌 새로운 기저 벡터가 있다고 했을 떄, 새로운 기저 벡터들의 좌표를 행렬 형태로 합쳐서 나타낸 것

# 좌표변화(coordinate transform)
### 새로운 기저 벡터에 대해 벡터 투영을 계산하는 것
### 새로운 기저 벡터에 대한 좌표 값이 가리키는 실제 위치는 원래의 벡터가 가리키는 실제 위치와 같아야 한다.

```python
e1 = np.array([1, 0])
e2 = np.array([0, 1])
a = np.array([2, 2])
g1 = np.array([1, 1])/np.sqrt(2)
g2 = np.array([-1, 1])/np.sqrt(2)

plt.annotate('', xy=e1, xytext=(0,0), arrowprops=dict(facecolor='green'))
plt.annotate('', xy=e2, xytext=(0,0), arrowprops=dict(facecolor='green'))
plt.annotate('', xy=g1, xytext=(0,0), arrowprops=dict(facecolor='red'))
plt.annotate('', xy=g2, xytext=(0,0), arrowprops=dict(facecolor='red'))
plt.annotate('', xy=a, xytext=(0,0), arrowprops=dict(facecolor='gray', alpha=0.5))

plt.plot(0, 0, 'ro', ms=10)
plt.plot(a[0], a[1], 'ro', ms=10)

plt.text(1.05, 1.35, "$a$", fontdict={"size": 18})
plt.text(-0.3, 0.5, "$e_2$", fontdict={"size": 18})
plt.text(0.5, -0.2, "$e_1$", fontdict={"size": 18})
plt.text(0.2, 0.5, "$g_1$", fontdict={"size": 18})
plt.text(-0.6, 0.2, "$g_2$", fontdict={"size": 18})

plt.xticks(np.arange(-2, 4))
plt.yticks(np.arange(-1, 4))
plt.xlim(-1.5, 3.5)
plt.ylim(-0.5, 3)
plt.grid(True)
plt.show()
```

```python
A = np.vstack([g1, g2]).T
A
```

```python
Ainv = np.linalg.inv(A)
Ainv
```

```python
Ainv.dot(a)
```
